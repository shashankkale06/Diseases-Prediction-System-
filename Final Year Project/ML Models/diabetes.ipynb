```python
df = pd.get_dummies(df,columns=['gender','smoking_history'])
df
```

```python
df.columns
```

```python
x = df[['age', 'hypertension', 'heart_disease', 'bmi', 'HbA1c_level',
       'blood_glucose_level','gender_Female', 'gender_Male',
       'smoking_history_No Info', 'smoking_history_current',
       'smoking_history_ever', 'smoking_history_former',
       'smoking_history_never', 'smoking_history_not current']]
y = df[['diabetes']]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(x_train.corr(), cmap="YlGnBu", annot=True)
```

```python
#Logisticregression
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
score1 = model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
# RandomForest
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=90, random_state=42, max_depth=9, max_features='sqrt', min_samples_split=15)
```

```python
model1.fit(x_train,y_train)
```

```python
score2 = model1.score(x_test,y_test)
```

```python
# svc
from sklearn.svm import SVC
from sklearn import svm
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[0.5,0.1,0.01],'C':[10,0.1,0.001,0.0001]}]
gsv = GridSearchCV(clf,param_grid,cv=10)
clf.fit(x_train,y_train)
```

```python
score3 = clf.score(x_test,y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt

models = ['Logistic Regression', 'Random Forest', 'SVC']
scores_test = [score1, score2, score3]

width = 0.35
x = np.arange(len(models))

fig, compare = plt.subplots(figsize=(8, 4))
bar1 = compare.bar(x, scores_test, width, label='Performance Data', color='blue')

compare.set_ylabel('Score')
compare.set_title('Model Performance Comparison')
compare.set_xticks(x)
compare.set_xticklabels(models)
compare.legend()
compare.set_ylim(0.3, 1.5)

for i, v in enumerate(scores_test):
    compare.text(i, v + 0.01, str(round(v, 2)), ha='center', va='bottom', fontsize=9)

plt.show()
```
